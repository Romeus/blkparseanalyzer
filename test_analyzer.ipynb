```python
# This file requires python 3.x (I used python 3.5) and the following packages installed: pandas, matplotlib
```

```python
# import needed functions
```

```python
from blkioanalyze.visualize import visualize_os_overhead, visualize_sector_reads, visualize_sector_counts
```

```python
from blkioanalyze.analyzer import BlkIOAnalyzer
```

```python
# x=0, y=0
```

```python
analyzer = BlkIOAnalyzer("samples/trace_x_0_y_0.bin")
```

```python
print("Minimal device response time %s" % analyzer.get_minimal_device_response_time())
```

```python
print("Average device response time %s" % analyzer.get_average_device_response_time())
```

```python
print("Maximum device response time %s" % analyzer.get_maximum_device_response_time())
```

```python
# visualise sector reads (sector/time)
```

```python
visualize_sector_reads(analyzer.get_sector_reads())
```

```python
# plot the graphic of os overheads
```

```python
visualize_os_overhead(analyzer.get_os_overhead())
```

```python
# visualize sector number of reads per sector
```

```python
visualize_sector_counts(analyzer.get_sectors_statistics())
```
